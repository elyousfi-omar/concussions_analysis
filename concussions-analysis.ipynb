Lets clean the dataset first, the date column should be of type datetime, not object

```python
data["Date"] = pd.to_datetime(data["Date"])
```

The Play Time After Injury and Average Playtime Before Injury columns are encoded as objects, let clean them, and convert them to numbers so we could check their distribution later.

```python
data["Play Time After Injury"] = data["Play Time After Injury"].replace("Did not return from injury", 0)
data["Play Time After Injury"].fillna(0, inplace=True)
data['Play Time After Injury'] = data['Play Time After Injury'].str.replace(r'\D+', '', regex=True).fillna(0)
data['Play Time After Injury'] = data['Play Time After Injury'].astype(int)
```

```python
data["Average Playtime Before Injury"].fillna(0, inplace=True)
data['Average Playtime Before Injury'] = data['Average Playtime Before Injury'].str.replace(r'\D+', '', regex=True).fillna(0)
data['Average Playtime Before Injury'] = data['Average Playtime Before Injury'].astype(int)
```

## Machine Learning

After analyzing data manually, and extracting patterns, now lets build a model to extract features importance, to see what affects the severity of the injury, we will measure this in number of games missed.

First, I have to encode categorical data.

```python
data.info()
```

```python
categories = ["Team", "Opposing Team", "Position", "Pre-Season Injury?", "Winning Team?", "Week of Injury", "Season", "Unknown Injury?", "Reported Injury Type"]
```

```python
y = data["Games Missed"]
data = data.drop(["Games Missed", "ID", "Player", "Date", "Game"], axis=1)
```

```python
data = pd.get_dummies(data, drop_first=True, columns=categories)
```

```python
data.head()
```

```python
x = data
```

```python
y.fillna(0, inplace=True)
```

```python
from sklearn import svm
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor, AdaBoostRegressor
from sklearn.model_selection import KFold, cross_val_score, GridSearchCV
from sklearn.metrics import recall_score, confusion_matrix, accuracy_score, classification_report, f1_score, roc_auc_score
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=43) 
```

Lets test multiple models.

```python
KFold_Score = pd.DataFrame()
classifiers = ['LinearRegression', 'RandomForestRegressor','GradientBoostingRegressor', 'AdaBoostRegressor']
models = [
          LinearRegression(),
          RandomForestRegressor(n_estimators=200, random_state=0),
          GradientBoostingRegressor(random_state=0),
          AdaBoostRegressor(random_state=0)
         ]
```

```python
j = 0
for i in models:
    model = i
    cv = KFold(n_splits=5, random_state=0, shuffle=True)
    KFold_Score[classifiers[j]] = -(cross_val_score(model, X_train, y_train, scoring = 'neg_mean_absolute_error', cv=cv))
    j = j+1
```

```python
mean = pd.DataFrame(KFold_Score.mean(), index= classifiers)
KFold_Score = pd.concat([KFold_Score,mean.T])
KFold_Score.index=['Fold 1','Fold 2','Fold 3','Fold 4', 'Fold 5', 'Mean']
KFold_Score.T.sort_values(by=['Mean'], ascending = True)
```

Random Forest Regressor, has the lowest mean absolute error. Lets use this model to predict number of games missed, using the features offered by the data.

```python
model = RandomForestRegressor(n_estimators=200, random_state=0)
model.fit(X_train, y_train)
```

### Findings

After building our machine learning model, that can predict number of games that a player can miss using the variables we have, what are the most influential features? In other words, what are the variables that affect number of missed games greatly?

```python
feature_importance = pd.DataFrame(data=model.feature_importances_, index=X_train.columns)
feature_importance.rename(columns={0: "Factors"}, inplace=True)
feature_importance.head()
```

```python
feature_importance = feature_importance["Factors"].nlargest(10)
```

```python
sns.barplot(y=feature_importance.index, x=feature_importance.values)

plt.xlabel("Importance")
plt.ylabel("Cause")
plt.title("What causes players that got head injuries to miss more games?")

plt.show()
```

* Interestingly, players that keep playing after getting injured, miss more games (Severe injuries!!). This is something that should be considered! Coaches should be aware that if a player gets a head injury in the game, it is better to sub them instantly, instead of waiting until the situation gets worse.
* Total snaps comes second as the reason of missing high number of games due to head injury. The snap count allows offensive players to have a small head start. The defensive players want to predict the snap, and build up speed such that they cross the line of scrimmage exactly as the play begins, to increase their chances of getting a tackle for a loss or a sack. (Wikipedia).
* Jacksonville Jaguars players, miss more games than any other team, due to head injury.
